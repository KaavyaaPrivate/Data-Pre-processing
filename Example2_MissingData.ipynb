There are four columns / features (three numerical - car, BuildingArea, YearBuilt and another categorical - councilarea) which are having missing values and need to be imputed. In the code below, an instance of SimpleImputer is created with strategy as "mean". The missing value is represented using NaN. Note some of the following:



1. sklearn.impute package is used for importing SimpleImputer class.
2. SimpleImputer takes two argument such as missing_values and strategy.
3. fit_transform method is invoked on the instance of SimpleImputer to impute the missing values 

````python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy="mean")
#we can fit the preprocessor with the car column
file.car = imp.fit_transform(file['Car'].values.reshape(-1,1))
#Output of car missing values
file
````

````python
file.BuildingArea = imp.fit_transform(file['BuildingArea'].values.reshape(-1,1))
file.YearBuilt = imp.fit_transform(file['YearBuilt'].values.reshape(-1,1))
#Output of BuildingArea, YearBuilt missing values.

#For example, from the output, we can cleary see the missing number of first row is filled(1121.8626243, 1121.8626243).
file
````

**Taking care of Categorical Features**

For handling categorical missing values, we should use one of the following strategies. However, it is the "most_frequent" strategy which is preferably used.

1. Most frequent (strategy='most_frequent')
2. Constant (strategy='constant', fill_value='someValue')
Here is how the code would look like when imputing missing value with strategy as most_frequent. In the code sample used in this post, councilarea is having missing values. Note how the missing value under councilarea column is replaced with 'Yarra' which occurs most frequently.

````python
from sklearn.impute import SimpleImputer

categorical_imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
file.CouncilArea = categorical_imputer.fit_transform(file['CouncilArea'].values.reshape(-1,1))
file
````
